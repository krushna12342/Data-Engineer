```python
from pyspark.sql import SparkSession
from pyspark.sql.types import *
from datetime import datetime, date

spark = SparkSession.builder.getOrCreate()

data = [
    (101, "Alice", "HR", 55000, {"Excel":"Advanced","Communication":"Expert"}, "New York", date(2024,1,10), datetime(2024,1,10,9,15,0)),
    (102, "Bob", "IT", 75000, {"Python":"Advanced","Spark":"Intermediate"}, "Chicago", date(2024,1,11), datetime(2024,1,11,10,20,0)),
    (103, "Charlie", "Finance", 68000, {"SQL":"Advanced","Excel":"Expert"}, "Dallas", date(2024,1,12), datetime(2024,1,12,11,30,0)),
    (104, "David", "IT", 82000, {"Java":"Expert","Kafka":"Intermediate"}, "Seattle", date(2024,1,13), datetime(2024,1,13,8,45,0)),
    (105, "Eva", "Marketing", 61000, {"SEO":"Advanced","Content":"Intermediate"}, "Boston", date(2024,1,14), datetime(2024,1,14,12,10,0)),
    (106, "Frank", "HR", 57000, {"Recruitment":"Expert","Excel":"Intermediate"}, "Miami", date(2024,1,15), datetime(2024,1,15,9,50,0)),
    (107, "Grace", "IT", 91000, {"Python":"Expert","AWS":"Advanced"}, "Austin", date(2024,1,16), datetime(2024,1,16,13,40,0)),
    (108, "Henry", "Finance", 70000, {"PowerBI":"Advanced","SQL":"Intermediate"}, "Denver", date(2024,1,17), datetime(2024,1,17,15,25,0)),
    (109, "Ivy", "Sales", 63000, {"CRM":"Advanced","Negotiation":"Expert"}, "Phoenix", date(2024,1,18), datetime(2024,1,18,16,5,0)),
    (110, "Jack", "IT", 85000, {"Scala":"Advanced","Spark":"Expert"}, "San Jose", date(2024,1,19), datetime(2024,1,19,17,35,0))
]

schema = StructType([
    StructField("emp_id", IntegerType(), True),
    StructField("name", StringType(), True),
    StructField("department", StringType(), True),
    StructField("salary", IntegerType(), True),
    StructField("skills", MapType(StringType(), StringType()), True),
    StructField("city", StringType(), True),
    StructField("date", DateType(), True),
    StructField("timestamp", TimestampType(), True)
])

df = spark.createDataFrame(data, schema)
df.display()
```
