```python
from pyspark.sql import SparkSession 
spark = SparkSession.builder.appName("Joins").getOrCreate() 
emp_df = spark.createDataFrame([ 
 (1, "John", 10), 
 (2, "Alice", 20), 
 (3, "Bob", 30), 
 (4, "David", 40) 
], ["emp_id", "emp_name", "dept_id"]) 
dept_df = spark.createDataFrame([ 
 (10, "HR"), 
 (20, "IT"), 
 (30, "Finance"), 
 (50, "Marketing") 
], ["dept_id", "dept_name"]) 
emp_df.show() 
dept_df.show() 
```

#self.join

```python
emp_df.join(dept_df,emp_df.dept_id == dept_df.dept_id,"inner" ).show()
```

```python
emp_df.join(dept_df,emp_df.dept_id == dept_df.dept_id,"left").show()
```

```python
from pyspark.sql.functions import col

emp = emp_df.alias("emp")
manager = emp_df.alias("manager")

emp.join(manager, col("emp.dept_id") == col("manager.dept_id"), "left").select(col("emp.emp_id"), col("emp.emp_name").alias("employee_name"), col("manager.emp_name").alias("manager_name")).show()
```
