#salary cotegery

```python

employees = [
("Ravi",120000),
("Amit",75000),
("John",35000)
]

schema = ["name","salary"]

df = spark.createDataFrame(employees,schema=schema)
df.show()
```

```python
def cotegory(salary):
    if salary > 100000:
        return "High"
    elif salary >50000:
        return "Medium" 
    else:
        return "Low"
    

cotegory_udf = udf(cotegory,StringType())
df_cotegory = df.withColumn("salary_cotegory",cotegory_udf(col("salary"))).show()
```

#Email Domain Extraction


```python
data= [(1,"vikas","vikas@gmail.com"),
       (2,"Ravi","ravi@gmail.com"),
       (3,"Amit","amit@gmail.com")]

df= spark.createDataFrame(data,schema=["id","name","email"])
```

```python
def domain (email):
    return email.split("@")[-1]

domain_udf=udf(domain,StringType())
df_email=df.withColumn("email",domain_udf("email")).show()
```

#SQL UDF

```python
def upper_name(name):
    return name.upper()

spark.udf.register("toUpper",upper_name,StringType())
df.createOrReplaceTempView("employee")

spark.sql("""select id ,name,toUPPER(name)from employee""").show()
```
